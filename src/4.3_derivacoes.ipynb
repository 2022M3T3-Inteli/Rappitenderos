# Agregação e Derivação das tabelas

*Tabela 2022-08-10 10_40am.csv exportada para selação dos quilômetros.*</span>

---



```python
df_km = pd.read_csv('/content/drive/Shareddrives/Grupo Rappitendeiros/2022-08-10 10_40am.csv', usecols=(['STOREKEEPER_ID', 'DISTANCE_TO_USER']), dtype={'STOREKEEPER_ID': 'float32', 'DISTANCE_TO_USER': 'float32'})

#código usado para retirar os valores NaN e transformar o 'STOREKEEPER_ID' em int32 para acelerar processamento

df_km = df_km.dropna().astype(dtype={'STOREKEEPER_ID': 'int32'}) 
```

*Criação da tabela df_km_sum a qual soma todos os kilometros de pedidos dos entregadores.* 

---

```python
#retirada dos valores nulos da coluna 'STOREKEEPER_ID' e a soma das características agupados por id.
df_km_sum = df_km.dropna(subset=['STOREKEEPER_ID']).groupby(by=['STOREKEEPER_ID']).sum()
df_km_sum
```

*Exportada tabela Ordens Done e df_cancel.csv para criação da tabela df_cancel_sum que é a quantidade de pedidos por entregador.*

---



```python
h=df_cancel = pd.read_csv('/content/drive/Shareddrives/Grupo Rappitendeiros/Ordens Done e Cancel.csv')
df_cancel_sum = df_cancel.dropna(subset=['STOREKEEPER_ID']).groupby(by=['STOREKEEPER_ID']).sum()
df_cancel_sum
#Nesse código ocorre a junção da tabela Ordens Done e df_cancel.csv para a criação da df_cancel_sum 
```

*Criação da tabela df_cancel_km que é o resultado da junção da df_cancel_sum e df_km_sum*

---



```python
df_cancel_km = df_km_sum.merge(df_cancel_sum, how = 'left', left_index=True, right_index=True)
df_cancel_km
#Nesse código representa a criação da tabela df_cancel_km que é a junção da  df_cancel_sum e df_km_sum
```

*Exportada tabela earnings.csv para a criação da tabela df_earnings_mean. df_earnings_mean é a media de renda por entregador.*

---



```python
 df_earnings=pd.read_csv("/content/drive/Shareddrives/Grupo Rappitendeiros/earnings.csv", usecols=(['STOREKEEPER_ID', 'EARNINGS','TIPS']), dtype={'STOREKEEPER_ID': 'float32', 'EARNINGS': 'float32'})    
 df_earnings= df_earnings.dropna().astype(dtype={'STOREKEEPER_ID': 'int32'})
 df_earnings_mean = df_earnings.dropna(subset=['STOREKEEPER_ID']).groupby(by=['STOREKEEPER_ID']).mean()
 df_earnings_mean
 #Nesse código representa a criação da tabela df_earnings_mean que é a média de receita do entregador
```

*Junção da tabela df_cancel_km com df_earnings_mean*

---



```python
df_cancel_km = df_cancel_km.merge(df_earnings_mean, how = 'left', left_index=True, right_index=True)
df_cancel_km
#Nesse código juntamos a tabela df_cancel_km com a tabela  df_earnings_mean
```

```python
df_supply = pd.read_csv('/content/drive/Shareddrives/Grupo Rappitendeiros/supply.csv', usecols=(['STOREKEEPER_ID', 'NUM_ORDERS','SUPPLY_HOURS']))
df_supply_mean = df_supply.dropna(subset=['STOREKEEPER_ID']).groupby(by=['STOREKEEPER_ID']).mean()
df_supply_mean
#Nesse código, limpamos a tabela e deixamos apenas as features "STOREKEEPER_ID","NUM_ORDERS"e "SUPPLY_HOURS"
```


Junção das tanbelas df_cancel_km e df_supply_mean

---



```python
df_cancel_km = df_cancel_km.merge(df_supply_mean, how = 'left', left_index=True, right_index=True)
df_cancel_km
#Nesse código, juntamos a tabela df_cancel com a tabela df_supply_mean
```

```python
from datetime import datetime, date
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month,  
                                                    born.day)) 
df_infos ['STOREKEEPER_YEAR']= df_infos ['DATA_NASCIMENTO'].apply(age)
df_infos 

#Código de agregação de features responsável por mostrar a idade do entregador("STOREKEEPER_YEAR") lendo sua data de nascimento
```
