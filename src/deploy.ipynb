## Processo de geração de modelo

Este código representa todas as etapas que nossa equipe realizou para chegar ao resultado que obtemos. 

```python
## Transformação dos arquivos .csv em dataframes do pandas e as colunas que iremos utilizar. 
### Aqui, utilizamos todos os arquivos inseridos na última etapa para transformá-los em dataframes do pandas, pegando apenas as colunas que nos são úteis e transformando os dados de 64bits para 32bits para carregar os dados mais rápido.

import pandas as pd
import numpy as np

df_gerais = pd.read_csv(gerais_csv, usecols=['ID'], dtype={'ID':'int32'})
df_churn = pd.read_csv(churn_csv, usecols=['ID'], dtype={'ID':'int32'})
df_earnings = pd.read_csv(earnings_csv, usecols=['STOREKEEPER_ID', 'EARNINGS'], dtype={'STOREKEEPER_ID':'int32', 'EARNINGS': 'float32'})
df_km = pd.read_csv(km_csv, usecols=['STOREKEEPER_ID', 'DISTANCE_TO_USER'], dtype={'STOREKEEPER_ID':'float32', 'DISTANCE_TO_USER': 'float32'})
df_resolucao = pd.read_csv(resolucao_csv, usecols=['STOREKEEPER_ID', 'TICKET_ID', 'RESPONSE_TIME'], dtype={'STOREKEEPER_ID':'int32', 'RESPONSE_TIME': 'int32'})
df_incidentes = pd.read_csv(incidentes_csv, usecols=['STOREKEEPER_ID', 'PUNISHMENT_TYPE', 'ORDER_ID'], dtype={'STOREKEEPER_ID':'int32', 'ORDER_ID': 'float32'})
df_orders = pd.read_csv(orders_csv, dtype={'STOREKEEPER_ID':'int32', 'ORDERS_DONE':'int32', 'ORDERS_CANCEL':'int32', 'CANCELS_OPS_RT': 'int32'})

## Derivações iniciais 
### Nesta seção, você verá todas as derivações necessárias das tabelas fonte para realizar a feature engineering que a nossa equipe chegou a conclusão de ser a mais eficiente para a predição. 

########## Target : 'CHURN' ##########

# Esta derivação tem como função servir de nossa categoria para dizer se um entregador deu ou não "CHURN", temos como base para essa derivação "Infos gerais" e "Criação contas churn".

# Primeiro, limparemos a tabela "Criação contas churn".

## CRIAÇÃO_CONTAS_CHURN ##

# Tabela "ID" salva em um dataframe específica.

df_churn_id = df_churn.drop(df_churn.columns.difference(['ID']), 1)

# Retirando os valores repetidos.

df_churn_id.drop_duplicates(inplace=True)

# Adição da coluna "Churn". 

df_churn_id['CHURN'] = True

# Neste ponto, nós adicionamos uma nova coluna chamada "CHURN", que será nosso target, pois, como todos estes ID's encontram-se na tabela "Criação contas churn", podemos assumir, com certeza, que todos esses ID's deram churn. 
# Agora, entraremos na tabela "INFOS GERAIS" para obter entregadores que nós sabemos que estão ativos. 

## INFOS GERAIS ## 

# Obtendo as informações "ID" da tabela "Infos gerais"

df_gerais_id = df_gerais.drop(df_gerais.columns.difference(['ID']), 1)

# Juntar as tabelas "df_gerais_id" e a tabela "df_churn_id"

df_ids_churn = df_churn_id.merge(df_gerais_id, how='outer', left_on='ID', right_on='ID')

# Preenchendo as tabelas "CHURN" que não são "1"

df_ids_churn['CHURN'].fillna(False, inplace=True)

# Agora, nós obtemos a tabela com todos os ids que tiveram churn e os ID's que nós temos certeza que estão ativos, concluindo a nossa coluna com o target. 

########## FIM DO TARGET ##########

########## TOTAL RESPONSE TIME ########## 

# Nesta derivação, nos obtemos todos os valores de espera para pedidos (tickets) de entregadores. Visando que quanto mais tempo eles esperam para ser resolvido seus problemas, maiores suas chances de churn. 

# Primeiro, escolhemos as colunas desejadas

df_resolucao_curated = df_resolucao.drop(df_resolucao.columns.difference(['TICKET_ID', 'RESPONSE_TIME']), 1).fillna(259200)

# Agora, iremos realizar a soma dos valores de 'RESPONSE_TIME'

df_total_response_time = df_resolucao_curated.groupby(['TICKET_ID']).sum().reset_index()

# Agora, devolveremos os valores do 'STOREKEEPER_ID' a tabela.

df_id_ticket = df_resolucao.drop(df_resolucao.columns.difference(['STOREKEEPER_ID', 'TICKET_ID']), 1)

# Removeremos os ID's duplicados.

df_id_ticket.drop_duplicates(subset='STOREKEEPER_ID')

# Finalmente, faremos a junção dos valores.

df_response_rt = df_total_response_time.merge(df_id_ticket, how='left', left_on='TICKET_ID', right_on='TICKET_ID')

# Remoção da coluna "TICKET_ID" e renomeação das colunas

df_response_rt = df_response_rt.drop('TICKET_ID', 1).rename(columns={'STOREKEEPER_ID': 'ID', 'RESPONSE_TIME': 'TOTAL_RESPONSE_TIME'})

# Remoção de colunas duplicadas

df_response_rt = df_response_rt.drop_duplicates(subset=['ID', 'TOTAL_RESPONSE_TIME'])

# Agrupando por ID para obter o valor total 

df_response_rt = df_response_rt.groupby(['ID']).sum().reset_index()

########## FIM DA TOTAL_RESPONSE_TIME ########## 

########## YEARLY_EARNINGS ##########
# Nesta feature, faremos a soma de todos os ganhos do entregador durante o ano, visando que quanto mais eles faturam, menor a chance de churn. 

# Soma dos valores das colunas "EARNINGS"

df_earnings_year = df_earnings.dropna(subset=['STOREKEEPER_ID']).groupby(['STOREKEEPER_ID']).sum().reset_index()

# Padronização dos nomes de coluna

df_earnings_year.rename(columns={'STOREKEEPER_ID': 'ID', 'EARNINGS': 'YEARLY_EARNINGS'}, inplace=True)

########## FIM DA YEARLY_EARNINGS ##########

########## DISTANCE_PER_ORDER #########
# Nesta feature, faremos a soma de todos os valores de distância (em KM) que o entregador percorreu durante uma fatia de tempo não determinada porém obtida pela soma das distâncias dos pedidos. 

#Média dos valores das colunas "DISTANCE_TO_USER"

df_quilometragem_average = df_km.groupby(by='STOREKEEPER_ID').mean().reset_index()

# Padronização dos nomes de coluna

df_quilometragem_average.rename(columns={'STOREKEEPER_ID': 'ID', 'DISTANCE_TO_USER': 'DISTANCE_PER_ORDER'}, inplace=True)

########## FIM DA DISTANCE_PER_ORDER ##########

########## ORDERS_DONE_RATE e TOTAL_ORDERS ##########
# Nesta feature, faremos a soma de todas os pedidos que o entregador teve em sua carreira, incluindo aqueles cancelados, e qual a média de conclusão dos pedidos que aquele entregador tem, visando que caso ele tenha muitos pedidos e baixa taxa de conclusão, este entregador está propenso a dar churn. 

# Criação da tabela "TOTAL_ORDERS"

df_orders['TOTAL_ORDERS'] = df_orders['ORDERS_DONE'] + df_orders['ORDERS_CANCEL'] + df_orders['CANCELS_OPS_RT']

# Criação da tabela "ORDERS_DONE_RATE"

df_orders['ORDERS_DONE_RATE'] = (df_orders['TOTAL_ORDERS'] - df_orders['CANCELS_OPS_RT'] - df_orders['ORDERS_CANCEL']) / df_orders['TOTAL_ORDERS']

# Remoção de valores "inf"

df_orders.replace([np.inf, -np.inf], 0, inplace=True)

# Isolação das colunas "ORDERS_DONE_RATE" e "TOTAL_ORDERS"

df_orders_totalrate = df_orders.drop(df_orders.columns.difference(['STOREKEEPER_ID', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS']), 1)

# Renomeação de colunas 

df_orders_totalrate.rename(columns={'STOREKEEPER_ID': 'ID'}, inplace=True)

########## FIM DA ORDERS_DONE_RATE E TOTAL_ORDERS ##########

########## RT_PUNISHED_ORDERS ##########
# Nesta feature, faremos a soma de todos os valores únicos de pedidos que sofreram qualquer tipo de punição para um entregador específico durante o período do ultimo ano, visando que entregadores que sofreram muitas punições estão mais propensos a dar churn e talvez não são desejáveis na plataforma.

# Obtenção da coluna "ORDER_ID"

df_incidentes_curated = df_incidentes.drop(df_incidentes.columns.difference(['STOREKEEPER_ID', 'ORDER_ID']), 1)

# Remoção de valores duplicados

df_incidentes_puntotal = df_incidentes_curated.drop_duplicates(subset=['ORDER_ID'])

# Substituição de valores 

df_incidentes_puntotal['ORDER_ID'] = 1

# Soma dos valores para entender o total de punimentos

df_incidentes_puntotal = df_incidentes_puntotal.dropna(subset=['STOREKEEPER_ID']).groupby(['STOREKEEPER_ID']).sum().reset_index()

# Renomeação das colunsa

df_incidentes_puntotal.rename(columns={'STOREKEEPER_ID': 'ID', 'ORDER_ID': 'RT_PUNISHED_ORDERS'}, inplace=True)

########## FIM DE RT_PUNISHED_ORDERS ##########

########## PERMANENT_BLOCK ##########
# Nesta feature, nós visamos ter controle daqueles entregadores que foram bloqueados permanentemente da plataforma, visando que entregadores que foram bloqueados permanentemente não são desejáveis a retornar mesmo que tenham churneado. 

# Obtenção das coluna "PUNISHMENT_TYPE"

df_incidentes_curated = df_incidentes.drop(df_incidentes.columns.difference(['STOREKEEPER_ID', 'PUNISHMENT_TYPE']), 1)

# Obtenção dos valores "permanent_block"

df_incidentes_permanent = df_incidentes_curated[df_incidentes_curated['PUNISHMENT_TYPE'] == 'permanent_block']

# Remoção dos ID's repetidos

df_incidentes_permanent.drop_duplicates(inplace=True)

# Renomeação da coluna

df_incidentes_permanent.rename(columns={'STOREKEEPER_ID': 'ID', 'PUNISHMENT_TYPE': 'PERMANENT_BLOCK'}, inplace=True)

# Ajuste das linhas

df_incidentes_permanent['PERMANENT_BLOCK'] = True

########## FIM DE PERMANENT_BLOCK ##########

### Agregação das features selecionadas
# Agora, depois de obtermos todas as features que julgamos imperativas para nossa visualização, realizaremos a junção (merge) das tabelas. 


df_all_features = df_ids_churn.merge(df_response_rt, how='left', left_on='ID', right_on='ID').merge(df_earnings_year, how='left', left_on='ID', right_on='ID').merge(df_quilometragem_average, how='left', left_on='ID', right_on='ID').merge(df_orders_totalrate, how='left', left_on='ID', right_on='ID').merge(df_incidentes_puntotal, how='left', left_on='ID', right_on='ID').fillna(0).merge(df_incidentes_permanent, how='left', left_on='ID', right_on='ID').fillna(False)


################## FIM DAS SELEÇÕES DAS FEATURES ##################

# Agora, faremos as seleções para realizar o nosso modelo.

# Declaração do Target
target = 'CHURN'

# Escolha do dataframe fonte
dataframe = df_all_features

# Substituição dos valores booleanos por valores inteiros
dataframe.replace({True: 1, False: 0}, inplace = True)

# Divisão entre X e Y

x = dataframe.drop(target, axis=1)

y = dataframe[target]

# Divisão entre X e Y de treino

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# Escalamento das features. 

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']] = scaler.fit_transform(x_train[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']])

x_test[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']] = scaler.transform(x_test[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']])

# Balanceamento do Target

from imblearn.under_sampling import RandomUnderSampler

ENN = RandomUnderSampler()
x_enn_train, y_enn_train = ENN.fit_resample(x_train, y_train)

# Definição das métricas

def metricas_enn(modelo):
  from sklearn.metrics import recall_score, precision_score, f1_score
  print('Acuracidade (treino): ', modelo.score(x_enn_train, y_enn_train))
  print('Acuracidade (teste): ', modelo.score(x_test, y_test))
  print('Revocação: ', recall_score(y_test, y_pred))
  print('Precisão: ', precision_score(y_test, y_pred))
  print('F1_score: ', f1_score(y_test, y_pred))


# Execução e hiperparametrização do modelo

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
params={
 "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
 "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
 "min_child_weight" : [ 1, 3, 5, 7 ],
 "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
 "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
}

xg=xgboost.XGBClassifier()

grid = RandomizedSearchCV(xg, param_distributions=params,scoring='roc_auc',n_jobs=-1,cv=5,verbose=3)

xgmodel = grid.fit(x_enn_train, y_enn_train)

xgmodel.best_estimator_

pd.DataFrame(x_train)

xgmodel = xgboost.XGBClassifier(colsample_bytree=0.7, gamma=0.0, learning_rate=0.05, max_depth=8)
xgmodel = xgmodel.fit(x_enn_train, y_enn_train)

y_pred = xgmodel.predict(x_test)
metricas_enn(xgmodel)

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Matriz de confusão do modelo

cm = confusion_matrix(y_test, y_pred, normalize = 'all')

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgmodel.classes_)
disp.plot()
plt.show()
```

```python
# Área de instalação de livrarias. 

!pip install featurewiz
!pip install catboost

# Área de importação de livrarias.

# Livrarias base

import pandas as pd 
import numpy as np 
import seaborn as sns

# Métricas e Visualização 

from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# Livrarias de Normalizadores

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import MinMaxScaler

# Livrarias de Balanceamentos

from imblearn.combine import SMOTEENN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE

# Livrarias de Algoritmos
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, ExtraTreesClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier 
import xgboost as xgb
from catboost import CatBoostClassifier
from sklearn.svm import SVC

# Livrarias de AutoML
from featurewiz import featurewiz


####### Definições e Dicionários ########

### Dicionários 

# Validação de escolha de Algoritmo 

algorithm_dict = {
    'KNN (K-Nearest Neighbors)': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': MultinomialNB(),
    'Random Forest': RandomForestClassifier(),
    'Regressão Logística': LogisticRegression(),
    'AdaBoost': AdaBoostClassifier(),
    'ExtraTrees': ExtraTreesClassifier(),
    'XGBoost': xgb.XGBClassifier(),
    'CatBoost': CatBoostClassifier(logging_level='Silent'),
    'BaggingClassifier': BaggingClassifier(),
    'SGD Classifier': SGDClassifier(),
    'SVM (Support Vector Machine)': SVC()
}

# Validação de escolha de Normalizador

scaler_dict = {
    'Standard Scaler': StandardScaler(),
    'MinMax Scaler': MinMaxScaler(),
    'MaxAbs Scaler': MaxAbsScaler()
}


# Validação de escolha de Balanceamento

balance_dict = {
    'Oversampling': SMOTE(),
    'Undersampling': RandomUnderSampler(),
    'Oversampling + Undersampling': SMOTEENN()
}

### Definições

def metricas(modelo):
  print('Acuracidade (treino): ', modelo.score(x_train, y_train))
  print('Acuracidade (teste): ', modelo.score(x_test, y_test))
  print('Revocação: ', recall_score(y_test, y_pred))
  print('Precisão: ', precision_score(y_test, y_pred))
  print('F1_score: ', f1_score(y_test, y_pred))
```

### Escolha de features automática

```python
# Importar o dataframe fonte: 
data = pd.read_csv(allfeatures_csv)

# Realizar avaliação das features: 

target = 'CHURN' #@param {type:"string"}
Algoritmo = 'Decision Tree' #@param ["KNN (K-Nearest Neighbors)", "Decision Tree", "Naive Bayes", "Random Forest", "Regressão Logística", "AdaBoost", "ExtraTrees", "XGBoost", "CatBoost", "BaggingClassifier", "SGD Classifier", "SVM (Support Vector Machine)"] {type:"string"}
Normalizador = 'Standard Scaler' #@param ["Standard Scaler", "MinMax Scaler", "MaxAbs Scaler"] {type:"string"}

# Separação entre X e Y

X = data.drop([target],axis=1)
 
y = data[(target)].values 

# Normalização das linhas

X_scaled =  scaler_dict[Normalizador].fit_transform(X) 

# Separação entre teste e treino

X_train, X_valid, y_train, y_valid = train_test_split(X_scaled,y,test_size = 0.2,stratify=y, random_state=1)

# Classificador base para avaliação de features

classifier = algorithm_dict[Algoritmo]
 
classifier.fit(X_train,y_train)

# Predição Inicial para avaliação

preds = classifier.predict(X_valid) 

pred1 = accuracy_score(preds,y_valid) 

# Seleção de features automática utilizando featurewiz

features, train = featurewiz(data, target, corr_limit=0.7, verbose=2, sep=",",
header=0,test_data="", feature_engg="", category_encoders="")

# Features escolhidas automaticamente

print(features)

# Realização de nova predição utilizando novos dados

X_new = train.drop(['CHURN'],axis=1)
 
y = train.CHURN.values 

# Normalização

X_scaled =  StandardScaler().fit_transform(X_new)

# Divisão entre treino e teste

X_train, X_valid, y_train, y_valid = train_test_split(X_scaled,y,test_size = 0.2,stratify=y, random_state=1)

# Instanciação do mesmo classificador para avaliação
 
classifier.fit(X_train,y_train)

# Resultado da Predição 


preds = classifier.predict(X_valid)

pred2 = accuracy_score(preds,y_valid) 

print()
print(f'Predição Inicial com Acurácia de {pred1}')
print()
print(f'As features escolhidas foram {features}')
print()
print(f'Predição Final com Acurácia de {pred2}')
```

### Gerar Novo Modelo

```python
# Importar o dataframe fonte: 
dataframe = pd.read_csv(allfeatures_csv)

# Agora, faremos as seleções para realizar o nosso modelo.

# Declaração do Target
target = 'CHURN' #@param {type:"string"}
Algoritmo = 'AdaBoost' #@param ["KNN (K-Nearest Neighbors)", "Decision Tree", "Naive Bayes", "Random Forest", "Regressão Logística", "AdaBoost", "ExtraTrees", "XGBoost", "CatBoost", "BaggingClassifier", "SGD Classifier", "SVM (Support Vector Machine)"] {type:"string"}
Normalizador = 'MinMax Scaler' #@param ["Standard Scaler", "MinMax Scaler", "MaxAbs Scaler"] {type:"string"}
Balanceamento = 'Undersampling' #@param ["Oversampling", "Undersampling", "Oversampling + Undersampling"] {type:"string"}
Hiperparametrização = 'Sem Hiperparametrização' #@param ["Sem Hiperparametrização", "Grid Search", "Random Search"] {type:"string"}


# Substituição dos valores booleanos por valores inteiros
dataframe.replace({True: 1, False: 0}, inplace = True)

# Divisão entre X e Y

x = dataframe.drop(target, axis=1)

y = dataframe[target]

# Divisão entre X e Y de treino

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# Escalamento das features. 

scaler = scaler_dict[Normalizador]

x_train[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']] = scaler.fit_transform(x_train[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']])

x_test[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']] = scaler.transform(x_test[['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']])

# Balanceamento do Target

balance = balance_dict[Balanceamento]
x_enn_train, y_enn_train = balance.fit_resample(x_train, y_train)

# Definição das métricas

def metricas_enn(modelo):
  from sklearn.metrics import recall_score, precision_score, f1_score
  print('Acuracidade (treino): ', modelo.score(x_enn_train, y_enn_train))
  print('Acuracidade (teste): ', modelo.score(x_test, y_test))
  print('Revocação: ', recall_score(y_test, y_pred))
  print('Precisão: ', precision_score(y_test, y_pred))
  print('F1_score: ', f1_score(y_test, y_pred))

pd.DataFrame(x_train)

xgmodel = algorithm_dict[Algoritmo]
xgmodel = xgmodel.fit(x_enn_train, y_enn_train)

y_pred = xgmodel.predict(x_test)

metricas_enn(xgmodel)

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# Matriz de confusão do modelo

cm = confusion_matrix(y_test, y_pred, normalize = 'all')

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgmodel.classes_)
disp.plot()
plt.show()
```

```python
df = df[['ID', 'TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS', 'PERMANENT_BLOCK', 'CHURN']]
df.replace({True: 1, False: 0,}, inplace = True)

df.drop(columns='CHURN', inplace=True)
df.head()

colunas = ['TOTAL_RESPONSE_TIME', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER', 'ORDERS_DONE_RATE', 'TOTAL_ORDERS', 'RT_PUNISHED_ORDERS']
df[colunas] = scaler.transform(df[colunas])
df.head()
df = df[['ID', 'TOTAL_RESPONSE_TIME', 'PERMANENT_BLOCK', 'RT_PUNISHED_ORDERS', 'TOTAL_ORDERS', 'ORDERS_DONE_RATE', 'YEARLY_EARNINGS', 'DISTANCE_PER_ORDER']]

xg_loaded.predict(df)

# df.to_excel('')# Template para transformar tabela em excel

pred = xg_loaded.predict(df)

df.head()
```

## Coluna de Ranqueamento

```python
from numpy.lib.function_base import append
lista = xg_loaded.predict_proba(df)
prob_results = []
for i in lista:
  if i[1] >= 0.8 and i[1] <= 1:
    prob_results.append("Churn")
 
  elif i[0] >= 0.85:
    prob_results.append("Nao_Churn")
  
  elif i[1] >= 0.65 and i[1] < 0.80:

    prob_results.append("Provavel_Churn")

  elif i[0] >= 0.65 and i[0] < 0.80 :
    prob_results.append("Provavel_Nao_Churn")
  else:
      prob_results.append("Incerto")

df['PRED_RANK'] = prob_results

df.head()

comp = df[['ID', 'PRED_RANK']]
comp['PRED'] = pred

comp
```
