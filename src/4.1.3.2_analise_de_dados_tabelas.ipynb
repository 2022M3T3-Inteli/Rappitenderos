# 1. Incidentes_Regras RT

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Cópia de Incidentes_Regras RT.csv')
df.head()
```

**Explicação:**

Analisando os conteúdos, é possível concluir que essa tabela é um método de controle das punições; ou seja, ela mostra pedidos que tiveram problemas, quem é o infrator, e qual a consequência dessa infração. 

**Descrição dos dados:**

* **Date** = Data da Infração 
* **Name** = Nome para manipulação interna de quais as punições possíveis de um RT 
* **Incident_ID** = Identificação do Incidente
* **Storekeeper_ID** = Identificação do RT
* **Punishment_Minutes** = Tempo de punição (em minutos)
* **Punishment_Type** = Tipo de punição 
* **Discipline_Rule_Bucket** = Tipos de classificações utilizadas para definir o tipo de advertência ou consequência que o RT terá. É uma derivada de um indicador interno da Rappi. 
* **Category_Rule** = Diz se é automático ou manual. 
* **Order_ID** = Identificação do Pedido
