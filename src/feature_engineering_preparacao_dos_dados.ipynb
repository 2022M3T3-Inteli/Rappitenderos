## 3. Cidade
Análise do churn de cada cidade separadamente 

```python
# seleção das colunas 'ID' e 'CIDADE' da tabela 'infos gerais.csv'
df_city = pd.read_csv('/content/drive/Shareddrives/Grupo Rappitendeiros/infos gerais.csv', usecols=['ID','CIDADE'], dtype={'ID':'int32'})
df_city = df_city.rename(columns={'CIDADE':'CITY'})
df_city.head()
```

```python
df_churn_city.drop(columns='ID')
```

```python
df_graph = df_churn_city.groupby(by=['CITY','CHURN']).count()
df_graph
```

```python
df_graph.reset_index()
```

```python
# seleciona as 10 cidades com maior número de RTs
df_graph.sort_values(['ID'], ascending=False).head(10).reset_index().CITY.unique()
```

```python
# cria uma lista com as 10 cidades
cidades = list(df_graph.sort_values(['ID'], ascending=False).head(10).reset_index().CITY.unique())
cidades
```

```python
# seleciona as 10 cidades dentro da tabela de churn por cidade
df_graph = df_graph.reset_index().query(f'CITY in {cidades}')
df_graph
```

```python
# contagem de churn true e false por cidade
df_graph = df_graph.rename(columns={'ID':'COUNT'})
df_graph
```

```python
import plotly.express as px

fig = px.bar(df_graph, x="CITY", y="COUNT", color="CHURN", title="Churn por cidade")
fig.show()
```

```python
# cria a coluna de total de RTs por cidade
df_total = df_graph.groupby(['CITY'])['COUNT'].sum().to_frame().reset_index()
df_total = df_total.rename(columns={'COUNT':'TOTAL'})
df_total
```

```python
    # adiciona a coluna do total do df do gráfico
df_graph = pd.merge(df_graph, df_total, how='right')
df_graph
```

```python
# cria a coluna de porcentagem de churn true e false por cidade
df_graph['PERCENTAGE'] = (df_graph['COUNT']/df_graph['TOTAL']).round(2)
df_graph
```

```python
# gráfico de porcentagem de churn por cidade
import plotly.express as px

fig = px.bar(df_graph, x="CITY", y="PERCENTAGE", color="CHURN", title="Churn por cidade")
fig.show()
```

Conclusão: o churn por cidade é uma constante, então não será utilizado como feature

# d) Identificação das features selecionadas, com descrição dos motivos de seleção.
As features serão adicionadas em apenas um tabela, armazenada na variável `null_churn`, criada a seguir

 Criação da tabela `null_churn` e inclusão das features `ID_ENTREGADOR`, `LEVEL_NAME` e `MODAL` da tabela `Product Return.csv`

```python
null_churn = pd.merge(target, df_productreturn, how='inner')
null_churn.head()
```

Inclusão das features `HAVE_CARD` da tabebela `df_cartao`  

```python
null_churn = pd.merge(null_churn, df_cartao, how='left')
null_churn.head()
```

```python
null_churn['CHURN'].value_counts()
```

Inclusão das features `COUNT_TO_GMV,` e `GMV` da tabela `df_gmv` 

```python
null_churn = pd.merge(null_churn, df_gmv, how='left')
null_churn.head()
```

```python
null_churn['CHURN'].value_counts()
```

Inclusão das features `PUNISHMENT_TYPE_permanent_block` `PUNISHMENT_TYPE_temporary_block` e `PUNISHMENT_TYPE_warning` da tabela `df_punishment`




```python
null_churn_merge = pd.merge(null_churn, df_punishment, how='left')
null_churn_merge.head(1)
```

Inclusão da tabela `out.csv`, originada a partir da elaboração de novas features



Junção das tabelas `null_churn_merge` e `out.csv` 

```python
df_out = pd.read_csv('/content/drive/Shareddrives/Grupo Rappitendeiros/out.csv', dtype={'ID':'int32'})
df_out.drop(df_out.columns[df_out.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df_out = df_out.drop(['TRANSPORTE'], axis=1)
df_out = df_out.rename(columns={'STOREKEEPER_ID':'ID'})
df_out = pd.merge(df_out, target, how='left')
df_out
```

```python
null_churn_final = pd.merge(null_churn_merge, df_out, how='left').fillna(0)
null_churn_final.head(1)
```
